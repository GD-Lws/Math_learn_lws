```python
%matplotlib inline
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
# 创建3d绘图区域
ax = plt.axes(projection='3d')

# 从三个维度构建
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

# 调用plot3D()绘制3d图形
#调用 ax.plot3D创建三维线图
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D line plot')
plt.show()
```

```python
# 3D散点图
fig = plt.figure()
#创建绘图区域
ax = plt.axes(projection='3d')
c = x + y
ax.scatter3D(x, y, z, c=c)
ax.set_title('3d Scatter plot')
plt.show()
```

<!-- 3D 等高线图 -->

```python
# 3D等高线图
def f(x, y):
       return np.sin(np.sqrt(x ** 2 + y ** 2))
#构建x、y数据
x_1 = np.linspace(-6, 6, 30)
y_1 = np.linspace(-6, 6, 30)
#将数据网格化处理
X, Y = np.meshgrid(x_1, y_1)
Z = f(X, Y)
fig_1 = plt.figure()
ax_1 = plt.axes(projection='3d')
#50表示在z轴方向等高线的高度层级，binary颜色从白色变成黑色
ax_1.contour3D(X, Y, Z, 50, cmap='binary')
ax_1.set_xlabel('x')
ax_1.set_ylabel('y')
ax_1.set_zlabel('z')
ax_1.set_title('3D contour')
plt.show()
```

```python
# 3D线框图
fig_2 = plt.figure()
ax_2 = plt.axes(projection='3d')
#调用绘制线框图的函数plot_wireframe()
ax_2.plot_wireframe(X, Y, Z, color='black')
ax_2.set_title('wireframe')
plt.show()
```

```python
# 3D曲面图
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
#矩阵转置
y = x.copy().T 
#数据z
z = np.cos(x ** 2 + y ** 2)
fig_3 = plt.figure()
ax_3 = plt.axes(projection='3d')
# 调用plot_surface()函数
ax_3.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax_3.set_title('Surface plot')
plt.show()
```

```python
import matplotlib.pyplot as plt


ax = plt.figure().add_subplot(projection='3d')

# Demo 1: zdir
zdirs = (None, 'x', 'y', 'z', (1, 1, 0), (1, 1, 1))
xs = (1, 4, 4, 9, 4, 1)
ys = (2, 5, 8, 10, 1, 2)
zs = (10, 3, 8, 9, 1, 8)

for zdir, x, y, z in zip(zdirs, xs, ys, zs):
    label = '(%d, %d, %d), dir=%s' % (x, y, z, zdir)
    ax.text(x, y, z, label, zdir)

# Demo 2: color
ax.text(9, 0, 0, "red", color='red')

# Demo 3: text2D
# Placement 0, 0 would be the bottom left, 1, 1 would be the top right.
ax.text2D(0.05, 0.95, "2D Text", transform=ax.transAxes)

# Tweaking display region and labels
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
