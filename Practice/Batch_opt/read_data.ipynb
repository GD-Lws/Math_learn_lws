```python
def data_tansform(file_path,bitch_num):
    raw_data = pd.read_csv(file_path)
    data_len = raw_data.shape[0]
    part_index_list = []
    X1_index_list = []
    for i in range(data_len):
        part_string = str(raw_data.iloc[i][2])
        if(part_string == "Part #"):
            part_index_list.append(i)
        elif(part_string == "X1"):
            X1_index_list.append(i)
    # print(part_index_list)
    # print(X1_index_list)
    part_num = len(part_index_list)
    data_num = 6
    df = pd.DataFrame(columns=['Bitch_num','Plate_id', 'item_id', 'x_coor', 'y_coor', 'length', 'width'])
    for i in range(part_num):
        now_part_index = part_index_list[i]
        X1_part_index = X1_index_list[i]
        for j in range(now_part_index + 2, X1_part_index - 1):
            df.loc[len(df.index)] = [bitch_num,i, raw_data.iloc[j][2], raw_data.iloc[j][12], raw_data.iloc[j][16], raw_data.iloc[j][19], raw_data.iloc[j][22]]
    return df
```
